```python
import pandas as pd
import numpy as np
import csv
import calendar
import datetime as dt
```

```python
senators_data = pd.read_csv('senators_data.csv')
```

```python
person = senators_data.drop_duplicates('senator').copy().dropna().reset_index().reset_index().drop(columns=['transaction_date','ticker','asset_type','type','amount','industry','index']).set_index('level_0')
person.index = np.arange(1, len(person) + 1)
```

```python
person=person.reset_index()
person.columns=['id','owner','state','person_name']
```

```python
ticker = senators_data.drop_duplicates('ticker').copy().dropna().reset_index().reset_index().drop(columns=['transaction_date','type','amount','state','owner','senator','index']).set_index('level_0')
ticker.index = np.arange(1, len(ticker) + 1)
```

```python
ticker=ticker.reset_index()
ticker.columns=['id','ticker','asset_type','industry']
```

```python
person_data = person.drop(columns='person_name')
```

```python
ticker_data = ticker.drop(columns='ticker')
```

```python
person=person.drop(columns=['owner','state'])
person = pd.concat([person, person_data], axis='columns', join='inner').drop(columns=['owner','state'])
```

```python
person.columns=['id','person_name','person_data']
```

```python
ticker=ticker.drop(columns=['asset_type','industry'])
ticker=pd.concat([ticker,ticker_data], axis='columns', join='inner').drop(columns=['asset_type','industry'])
ticker.columns=['id','ticker','ticker_data']
```
